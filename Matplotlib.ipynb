```python
import numpy as np
import matplotlib.pyplot as plt
```

ข้อที่ 1

```python
t = np.arange(0, 11, 1)  # เวลา 0 ถึง 10 เพิ่มทีละ 1
x = 2*t + 1              # ความสัมพันธ์

plt.plot(t, x)
plt.xlabel("time (t)")
plt.ylabel("x = 2t + 1")
plt.title("The relationship between x and t")
plt.grid()
plt.show()
```

ข้อที่ 2

```python
m = 1.0
v = np.array([0, 1, 2, 3, 4, 5])
Ek = 0.5 * m * v**2

plt.plot(v, Ek, marker='o')
plt.xlabel("v (m/s)")
plt.ylabel("Ek (J)")
plt.title("Graph of kinetic energy versus velocity")
plt.grid()
plt.show()
```

ข้อที่ 3

```python
T = np.array([22.1, 23.4, 25.0, 27.5, 29.2, 30.0, 29.0, 27.0])

t = np.arange(len(T))

Tmax = np.max(T)
tmax = np.argmax(T)

print("Tmax =", Tmax)
print("Time =", tmax)

plt.plot(t, T, marker='o')
plt.plot(tmax, Tmax, 'ro', label="max")
plt.xlabel("Time (hr)")
plt.ylabel("Temperature(°C)")
plt.legend()
plt.grid()
plt.show()
```

ข้อที่ 4

```python
x = np.linspace(0, 2*np.pi, 200)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="cos(x)")
plt.xlabel("x")
plt.ylabel("Value")
plt.title("Sine and cosine graphs")
plt.legend()
plt.grid()
plt.show()
```

ข้อ 5

```python
a = 2.0
v0 = 1.0

t = np.linspace(0, 5, 100)
v = v0 + a*t

plt.plot(t, v)
plt.xlabel("t (s)")
plt.ylabel("v (m/s)")
plt.title("Velocity over time v = v0 + at")
plt.grid()
plt.show()
```

ข้อ 6

```python
A = 2.0
omega = np.pi

t = np.arange(0, 2, 0.01)
x = A * np.sin(omega * t)

plt.plot(t, x)
plt.xlabel("t (s)")
plt.ylabel("x(t)")
plt.title("Simple harmonic motion")
plt.grid()
plt.axhline(0, color='black')
plt.show()
```

ข้อ 7

```python
h = 6.626e-34     # แผลงค์
c = 3e8           # ความเร็วแสง
k = 1.38e-23      # ค่าคงที่ Boltzmann
T = 5000          # อุณหภูมิ

lam = np.linspace(100e-9, 3000e-9, 500)  # 100–3000 nm → เมตร

I = (2*h*c**2) / (lam**5 * (np.exp((h*c)/(lam*k*T)) - 1))

plt.plot(lam*1e9, I)   # แปลงกลับเป็นหน่วย nm
plt.xlabel("λ (nm)")
plt.ylabel("Intensity (a.u.)")
plt.title("Blackbody Radiation (Planck's Law)")
plt.grid()
plt.show()
```

ข้อ 8

```python
Q = 1e-6
k = 9e9

r = np.linspace(0.1, 1.0, 100)
E = k*Q / r**2

plt.plot(r, E)
plt.xlabel("r (m)")
plt.ylabel("E (N/C)")
plt.title("Electric field from a point charge")
plt.grid()
plt.show()
```

ข้อ 9

```python
# ค่าคงที่
G = 6.67e-11           # J m/kg^2
M = 5.97e24            # kg
m = 1000               # kg

# r ตั้งแต่ 6400 km ถึง 20000 km (แปลงเป็นเมตร)
r_km = np.linspace(6400, 20000, 300)
r = r_km * 1000    # เมตร

# พลังงานศักย์โน้มถ่วง
U = G * M * m / r

plt.plot(r_km, U)
plt.xlabel("r (km)")
plt.ylabel("U (J)")
plt.title("Gravitational potential energy U versus r")
plt.grid()
plt.show()
```

ข้อ 10

```python
# ค่าคงที่
m0 = 1.0
c = 3e8

# ความเร็ว v = 0 → 0.999c เพิ่มทีละ 0.05c
v = np.arange(0, 0.999, 0.05) * c

# คำนวณมวลสัมพัทธภาพ
m = m0 / np.sqrt(1 - (v**2 / c**2))

ratio = m / m0       # m/m0
beta = v / c         # v/c

plt.plot(beta, ratio, marker='o')
plt.xlabel("v/c")
plt.ylabel("m / m0")
plt.title("The increase of relativistic mass")
plt.grid()
plt.show()
```
